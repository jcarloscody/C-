```python
**********************************   FUNÇÃO AFIM ****************************************

f(x) = ax + b

  a -> declividade ou coeficiente angular em relação ao eixo horizontal. representa a taxa de crescimento ou taxa de variação da função.

  b -> valor inicial da função f ou coeficiente linear da reta.

  x -> variável independente

PROPRIEDADES 
  -> Uma função afim é crescente quando sua taxa de variação a é positiva, e decrescente quando a taxa de variação a é negativa.

TIPOS 
  -> ascendente: a > 0 
  -> descendente: a < 0 
  -> linear: f(x) = ax, a reta passa pela origem (0,0) 
  -> identidade: f(x) = x, a=1, o gráfico é a bissetriz de um ângulo. 
  -> bissetriz: É a semirreta que parte do vértice do ângulo e o divide em dois ângulos de mesma medida 
  -> constante: f(x) = b, é paralela ao eixo da abscissas
```

```python
import matplotlib.pyplot as plt
xAscendente = [-2,-1,0,1,2]
yAscendente = [-3,-1,1,3,5]
plt.axvline(0)
plt.axhline(0)
plt.style.use('ggplot')
plt.title("Gráfico Função Afim: ascendente     f(x)= 2x + 1")
plt.xlabel("Eixo das Abscissas")
plt.ylabel("Eixo das Coordenadas")
plt.plot(xAscendente,yAscendente)
plt.text(0,1,"f(0)=b=1", fontsize=12,horizontalalignment="right")
plt.text(0,1,"**  ponto(0,1) ponto em que a reta intersecta o eixo y", fontsize=12,horizontalalignment="left")

plt.show()
```

```python
xdescendente = [-1,0,1]
ydescendente = [5,2,-1]
plt.axvline(0)
plt.axhline(0)
plt.style.use('ggplot')
plt.title("Gráfico Função Afim: descendente    f(x) = -3x + 2")
plt.xlabel("Eixo das Abscissas")
plt.ylabel("Eixo das Coordenadas")
plt.plot(xdescendente,ydescendente)
plt.text(0,2,"f(x) = b = 2     * ", fontsize=12,horizontalalignment="right")
plt.text(0,2,"*  ponto(0, 2) ponto em que a reta intersecta o eixo y", fontsize=12,horizontalalignment="left")

plt.show()
```

```python
************************************** FUNÇÃO QUADRÁTICA *****************************

    ->  também chamada de função polinomial de 2º grau
    -> ax**2 + bx + c      Onde a, b e c são números reais e a ≠ 0.
    
    -> Atenção! 
    Observe que não são funções quadráticas:
        -> f(x) = 2x   É função linear.
        -> f(x) = 2**x  É função exponencial.
        -> f(x) = x**3 + 2x**2 + x + 1   É função do 3o grau.

-> fato histórico:
    -> há 4mil anos (babilônios)
    -> Muitos dos problemas que foram encontrados nos tabletes dos babilônios consistiam em determinar dois números conhecendo a soma e o produto deles; 
    ou em encontrar os lados de um retângulo conhecendo o perímetro e a área.
    -> Desde cerca de 1800a.C. até o século XVI todos os matemáticos resolviam as equações do 2o grau seguindo as “receitas” iguais ou semelhantes 
    às dos antigos babilônios.

RAÍZES
    -> x2**2 + bx + c = 0
    -> determinar os valores de x para os quais a função se anula
    -> Esses valores são chamados zeros da função quadrática ou raízes


OBSERVAÇÃO:
    -> Dados quaisquer b e c,nem sempre existem dois números reais cuja soma seja b e cujo produto seja c.
        -> Por exemplo, não existem dois números reais cuja soma seja 3 e cujo produto seja 7.
    -> O número Delta é chamado discriminante da função quadrática
    -> Quando Delta > 0, a função tem dois zeros reais diferentes.
        -> Delta < 0  não tem zeros.
    -> Quando Delta = 0, tem dois zeros reais iguais.

CONCAVIDADE:
    -> quando a > 0 a concavidade para cima.
    -> quando a < 0,a concavidade para baixo.
    -> maior o valor absoluto de a,menor será a abertura da parábola(parábola mais “fechada”), independentemente da concavidade ser para 
    cima ou para baixo.
    -> Se b > 0,a parábola intersecta o eixo y no ramo crescente
    -> Se b < 0, a parábola intersecta o eixo y no ramo decrescente.
    -> Se b = 0, a parábola intersecta o eixo y no vértice. pag 119
```

```python
*************************  FUNÇÃO LOGARÍTMICA   ******************************************   


Lei f(x) = logax=c

    ->  A função inversa da função logarítmica é a função exponencial.
    -> "a" é a base de logaritmo
    -> a > 0 && a <> 1
    -> x > 0 && x E Reais
    -> "c" é o logaritmo
    -> "x" é o logaritmando
    -> o gráfico nunca toca o eixo y nem ocupa pontos dos quadrantes II e III;
    -> Injetora e sobrejetora
    
TIPOS:
    -> crescente: a > 0
    -> decrescente: 0 < a < 1
    
PROPRIEDADES
    -> log a(M * N) = log aM + log aN
    -> Log a(M/N) = Log aM - Log aN
    -> log aM**N = N * log aM
    

FATO HISTÓRICO
    -> As antigas tabelas de cossenos tinham, naturalmente, pouca precisão para as exigências do século XVII, e a invenção dos logaritmos veio 
    substituir com enorme vantagem as Fórmulas de Prostaférese.
    -> A teoria dos logaritmos proporcionou o desenvolvimento de uma metodologia para a construção de tabelas com grande precisão, 
    que passaram a conter os logaritmos dos senos, cossenos e tangentes. A nova tecnologia foi imediatamente adotada e mesmo os cálculos mais complicados 
    já tinham à disposição tabelas com até catorze decimais.
    -> cosX * cosY = 0.5[cos(x + y) + cos(x - y)] Fórmulas de Prostaférese   pag 189
    
```

```python
xLogaritmaCrescente = [1/4,0.5,1,2,4]
yLogaritmaCrescente = [-2,-1,0,1,2]
plt.axvline(0)
plt.axhline(0)
plt.style.use('ggplot')
plt.title("Gráfico Função Logarítma Crescente   f(x) = log 2X    a > 1")
plt.xlabel("Eixo das Abscissas")
plt.ylabel("Eixo das Coordenadas")
plt.plot(xLogaritmaCrescente,yLogaritmaCrescente)
plt.show()
```

```python
**************************** FUNÇÃO EXPONENCIAL ***************************************

lei de formação da função exponencial   f(x) =a**x    
    
    -> Função Exponencial é aquela que a variável está no expoente e cuja base é sempre maior que zero e diferente de um.
    -> a (base): é um número real positivo diferente de 1
    -> função exponencial sempre estará no I e II quadrantes
    
    ->  Para que essa função? crescimento populacional, evolução de capital por juros compostos, decaimento radioativo de substâncias químicas,
     desenvolvimento de bactérias em uma colônia, comportamento de novos casos da covid-19

TIPOS
    -> crescente: a > 1
    -> decrescente: 0 < a < 1

-> OBSERVAÇÕES pag 161
```
